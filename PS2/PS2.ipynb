```python
import math
import matplotlib.pyplot as plt
```

```python
print(math.factorial(100))

print(len(str(math.factorial(100))), " digits")
```

```python
x = range(51)

plt.plot(x, [0]+[i*math.log(i)-i for i in x[1:]], label='Stirling')
plt.plot(x, [math.log(math.factorial(i)) for i in x], label='actual')
plt.legend()
plt.show()
```

```python
plt.plot(x, [1]+[(math.factorial(i))/(math.e**(i*math.log(i)-i)) for i in x[1:]])
plt.show()
```
